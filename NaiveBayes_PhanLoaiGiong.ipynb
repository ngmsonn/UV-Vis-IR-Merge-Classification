```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, r2_score, confusion_matrix, ConfusionMatrixDisplay, precision_score, f1_score, recall_score
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import LinearSVC, SVC
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from scipy.signal import savgol_filter
from icecream import ic
import random
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv("/home/sonn/Son/Workspace/UV-Vis_IR_NuocCam/data/data_tunhien.csv")
data.head()
```

```python
y = data['Giong']
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X = data.iloc[:,4:]
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

GaussianNB_KhongPCA

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix(y_test, clf.predict(X_test)))
fig, ax = plt.subplots(figsize=(6,5))
cmd.plot(ax=ax)
plt.show()
```

GaussianNB_PCA

```python
pca = PCA()
clf = GaussianNB()
pipeline = Pipeline(steps=[('pca', pca), ('clf', clf)])
param_grid = {
    'pca__n_components': [2,3,4,5,6,7,8,9,10,12,14],
}
grid = GridSearchCV(pipeline, param_grid, refit = True, verbose = 3) 
grid.fit(X_train, y_train) 
print("\n\n")
ic(grid.best_params_) 
model = grid.best_estimator_
model.fit(X_train, y_train)
print("Accuracy score:",accuracy_score(y_test, model.predict(X_test)))
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix(y_test, model.predict(X_test)))
fig, ax = plt.subplots(figsize=(6,5))
cmd.plot(ax=ax)
plt.show()
```

BernoulliNB

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
clf = BernoulliNB()
clf.fit(X_train, y_train)
accuracy_score(y_test, clf.predict(X_test))
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix(y_test, clf.predict(X_test)))
fig, ax = plt.subplots(figsize=(6,5))
cmd.plot(ax=ax)
plt.show()
```

```python
pca = PCA()
clf = BernoulliNB()
pipeline = Pipeline(steps=[('pca', pca), ('clf', clf)])
param_grid = {
    'pca__n_components': [2,3,4,5,6,7,8,9,10,12,14],
}
grid = GridSearchCV(pipeline, param_grid, refit = True, verbose = 3) 
grid.fit(X_train, y_train) 
print("\n\n")
ic(grid.best_params_) 
model = grid.best_estimator_
model.fit(X_train, y_train)
print("Accuracy score:",accuracy_score(y_test, model.predict(X_test)))
```
